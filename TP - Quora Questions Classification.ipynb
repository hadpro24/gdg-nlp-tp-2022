# 3 - Processing

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords, wordnet
from nltk import pos_tag
from sklearn.feature_extraction.text import CountVectorizer
import re
import numpy as np
def get_wordnet_pos(pos_tag):
    output = np.asarray(pos_tag)
    for i in range(len(pos_tag)):
        if pos_tag[i][1].startswith('J'):
            output[i][1] = wordnet.ADJ
        elif pos_tag[i][1].startswith('V'):
            output[i][1] = wordnet.VERB
        elif pos_tag[i][1].startswith('R'):
            output[i][1] = wordnet.ADV
        else:
            output[i][1] = wordnet.NOUN
    return output

def processing(question):
    question = re.sub(r'\S+@[\w-]+.\w{2,3}', '', question)
    question = re.sub(r'(https|http)\S+', '', question)
    question = re.sub("<[^>]*>", "", question)
    question = re.sub(r"@\S+", "", question)
    question = re.sub(r"#\S+", "", question)
    tokens = word_tokenize(question)
    tokens = [word.lower() for word in tokens if word.isalpha()]
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    pos_taging = pos_tag(tokens)
    lemma = WordNetLemmatizer()
    tokens = [lemma.lemmatize(word, tag) for word, tag in get_wordnet_pos(pos_taging)]
    return tokens
```

```python
df_sample['tokens'] = df_sample['question_text'].apply(processing)
```

```python
df_sample.head()
```

## 4 - Sentiment analysis 

```python
from textblob import TextBlob
```

```python
def polarity(doc):
    return TextBlob(doc).sentiment.polarity

def subjectivity(doc):
    return TextBlob(doc).sentiment.subjectivity
```

```python
df_sample['polarity'] = df_sample['question_text'].apply(polarity)
```

```python
df_sample['subjectivity'] = df_sample['question_text'].apply(subjectivity)
```

```python
df_sample.head()
```

# 5 - Word embedding

```python
import numpy as np
with open("../d8/glove.6B/glove.6B.50d.txt", encoding='utf-8') as file:
    embedding_words = {}
    for line in file.readlines():
        word_vector = line.split()
        word = word_vector[0]
        vector = word_vector[1:]
        embedding_words[word] = np.asarray(vector, dtype='float32')
```

```python
len(embedding_words)
```

```python
embedding_words['is']
```

```python
def embed(tokens):
    return np.mean([np.mean(embedding_words.get(token, 0.0)) for token in tokens] or 0.0)
```

```python
df_sample['embedding'] = df_sample['tokens'].apply(embed)
```

```python
df_sample.head()
```

# 6 - Vectorization

```python
vectorizer = CountVectorizer(analyzer=lambda x:x)
vectors = vectorizer.fit_transform(df_sample['tokens']).toarray()
```

```python
df_bow = pd.DataFrame(data=vectors, columns=vectorizer.get_feature_names())
df_bow.head()
```

```python
df_bow.shape
```

```python
df_bow['polarity'] = df_sample['polarity'].tolist()
df_bow['subjectivity'] = df_sample['subjectivity'].tolist()
df_bow['embedding'] = df_sample['embedding'].tolist()
```

```python
df_bow
```

# 7 - Train model

```python
X = df_bow
y = df_sample['target']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
ypred = model.predict(X_test)
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
```

```python
f1_score(y_test, ypred)
```

```python
print(classification_report(y_test, ypred))
```
